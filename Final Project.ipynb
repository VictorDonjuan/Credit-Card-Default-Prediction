## The data

The dataset that we're going to use is public and can be downloaded from here [UCI Machine Learning Repository](http://archive.ics.uci.edu/ml/datasets/default+of+credit+card+clients).

The data comes from a credit card company from Taiwan and includes some demographics and recent financial data over 6 months (from April 2005 to September 2005).

This dataset consists of 30000 observations with 25 columns in total, from which 23 are features since a column was made for an ID number, and naturally we have a response variable, namely "default payment next month". This is a label for every row that tells us whether or not, in the next month after the 6-month historical data, an account owner has defaulted, which means that they didn't make the minimum payment (in NT dollars, i.e., New Taiwan dollars).

The objective of this notebook is to create a model that can effectively predict whether or not an account will default next month, given the historical data.

After we've created the model, we will later use these predictions in order to perform a Financial Analysis to estimate the savings if we prevented accounts from defaulting.

```python
df = pd.read_excel('default of credit card clients.xls')
df.head()
```

The information of every feature is described from the source link, and it is as follows.

Attribute Information:

This research employed a binary variable, default payment (Yes = 1, No = 0), as the response variable. This study reviewed the literature and used the following 23 variables as explanatory variables:

X1: Amount of the given credit (NT dollar): it includes both the individual consumer credit and his/her family (supplementary) credit.

X2: Gender (1 = male; 2 = female).

X3: Education (1 = graduate school; 2 = university; 3 = high school; 4 = others).

X4: Marital status (1 = married; 2 = single; 3 = others).

X5: Age (year).

X6 - X11: History of past payment. We tracked the past monthly payment records (from April to September, 2005) as follows: X6 = the repayment status in September, 2005; X7 = the repayment status in August, 2005; . . .;X11 = the repayment status in April, 2005. The measurement scale for the repayment status is: -1 = pay duly; 1 = payment delay for one month; 2 = payment delay for two months; . . .; 8 = payment delay for eight months; 9 = payment delay for nine months and above.

X12-X17: Amount of bill statement (NT dollar). X12 = amount of bill statement in September, 2005; X13 = amount of bill statement in August, 2005; . . .; X17 = amount of bill statement in April, 2005.

X18-X23: Amount of previous payment (NT dollar). X18 = amount paid in September, 2005; X19 = amount paid in August, 2005; . . .;X23 = amount paid in April, 2005.

To make it clearer, we will use the first row as our column names, and let's also rename 'PAY_0' to 'PAY_1' for consistency.

```python
# first row as column names
df.columns = df.iloc[0]

# drop the first row
df.drop(0, inplace=True)

# rename column PAY_0 to PAY_1 
df.rename(columns={'PAY_0':'PAY_1'}, inplace=True)
```

## Verifying integrity of the data

Let's check if we don't have missing values, and the general estructure of our dataset.

```python
df.info()
```

To begin with, it seems like a good idea to drop the ID column since it's the same as our index.

Next, since all of our columns are stored as strings ('object'), we should convert those to int numbers, except for the categorical features. The categorical features should be 'SEX', 'EDUCATION', and 'MARRIAGE' since they can't really be ordered even if they were assigned numerical values. Also, we will leave the response variable as is.

```python
# drop the ID column
df.drop('ID', axis=1, inplace=True)

# separating columns between categorical and numerical
all_columns = df.columns
cat_columns = ['SEX', 'EDUCATION', 'MARRIAGE']
num_columns = [column for column in all_columns if column not in cat_columns]
```

```python
# convert strings from numerical columns to int numbers
df[num_columns] = df[num_columns].astype(int)

# checking if the changes were correctly made
df.info()
```

Next, there's a detail about the 'PAY_N' features that should be adressed. For example, let's take a look at all the unique values from 'PAY_1'.

```python
df['PAY_1'].unique()
```

The problem with this is that we have two undocumented values, namely -2 and 0. Let's recall what information we did have:

The measurement scale for the repayment status is: -1 = pay duly; 1 = payment delay for one month; 2 = payment delay for two months; . . .; 8 = payment delay for eight months; 9 = payment delay for nine months and above.

After further researching on this, the following explanation was provided:

-2 means the account started that month with a zero balance and never used
any credit.

0 means that at least the minimum payment was made, but the entire balance
wasn't paid (that is, a positive balance was carried to the next month).

We will explore now the rest of the features.

The first one of the financial features we want to explore is the 'PAY_N', where N = 1,...,6. Let's remember what these values mean from the attribute information:

The values are describes like this: -1 = pay duly; 1 = payment delay for one month; 2 = payment delay for two months; . . .; 8 = payment delay for eight months; 9 = payment delay for nine months and above. -2 means the account started that month with a zero balance and never used any credit. 0 means that at least the minimum payment was made, but the entire balance wasn't paid (that is, a positive balance was carried to the next month).

The 'N' in 'PAY_N' refers to the corresponding month: 1 = September, 2 = August, 3 = July, 4 = June, 5 = May, 6 = April.

Let's see the distribution of these 6 features.

```python
# defining bin edges 
# so that the center of the bins are integers
pay_bins = np.array(range(-2,10)) - 0.5

# plotting histograms for 'PAY_N' (N from 1 to 6) 
fig, axes = plt.subplots(3, 2, dpi=200, figsize=(12,12))

fig.suptitle('Distribution of PAY_N features', size=18)

sns.histplot(ax=axes[0,0], data=df, x='PAY_1', bins=pay_bins)
sns.histplot(ax=axes[0,1], data=df, x='PAY_2', bins=pay_bins)
sns.histplot(ax=axes[1,0], data=df, x='PAY_3', bins=pay_bins)
sns.histplot(ax=axes[1,1], data=df, x='PAY_4', bins=pay_bins)
sns.histplot(ax=axes[2,0], data=df, x='PAY_5', bins=pay_bins)
sns.histplot(ax=axes[2,1], data=df, x='PAY_6', bins=pay_bins)

fig.subplots_adjust(wspace=.2, hspace=.3);
```

Here we run into a difficult problematic, because it's clear the a lot of 1's are missing from PAY_2 to PAY_6. For instance, if a certain amount of people in August have a PAY_2 equal to '2' (they haven't paid for two months), it can logically only mean that they must have had a PAY_3 (July) equal to '1' (they haven't paid for one month). But this is not the case according to the histograms. This is a clear sign of wrong imputation in the data, and the only choice we have here is to drop these 5 features:

```python
df.drop(['PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6'], axis=1, inplace=True)
```

Now that the data is cleaned, we can look at a heatmap of the feature correlation numbers.

```python
# column name shortened
temp_df = df.rename(columns={'default payment next month':'default next m.'})

# heatmap of the correlations
plt.figure(dpi=200, figsize=(10,7))
sns.heatmap(round(temp_df.corr(),2), annot=True, cmap='viridis');
```

As we can see, the only features that have slitghly significant correlation with defaulting next month are LIMIT_BAL (as we saw before) and PAY_1. This makes sense. If we one-hot encode the categorical features, we can see their numeracal value for the correlation between them and the label.

```python
# one-hot encoding categorical features
onehot_cat_features = pd.get_dummies(df[cat_columns])

# concatenating with the response variable
temp_df = pd.concat([onehot_cat_features, df['default payment next month']], axis=1)

# resulting heatmap of the correlations
plt.figure(dpi=200, figsize=(10,7))
sns.heatmap(temp_df.corr(), annot=True);
```

As expected from the previous analyses, no clear correlation with one of the categorical features was found.

## Preparing Data for Modelling

To finish this part of the project, let's prepare our data to be trained and tested on. First, we will get the one-hot encoded categorical features along with the rest of the features as our set 'X', and then the response variable will be stored as 'y'.

Also, it is important to note that since we are planning on using a XGBoost Classifier as our model, we don't need to scale the features because tree-based models don't work with (Euclidean) distances. 

```python
X = pd.get_dummies(df.drop('default payment next month', axis = 1))
y = df['default payment next month']
```

Next, we will choose 80% of the dataset as our train data. We will perform and random split with sklearn's function train_test_split and a random_state equal to 24.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=24)
```
